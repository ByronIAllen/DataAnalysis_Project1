```python
new_cdc = cdc_data [["YearEnd","LocationAbbr","Question","DataValue","DataValueType","StratificationCategory1"]]
new_cdc

Lack_insurance = new_cdc.loc[new_cdc["Question"] == "Current lack of health insurance among adults aged 18-64 years", :]
Lack_insurance.head()

```

```python
Overall_Lack = Lack_insurance.loc[Lack_insurance["StratificationCategory1"] == "Overall",:]
Overall_Lack.head()
```

```python
Insurance_Drop = Overall_Lack.dropna()
Insurance_Drop.head()
```

```python
Crude_Insurance = Insurance_Drop.loc[Insurance_Drop["DataValueType"] == "Crude Prevalence",:]
Crude_Insurance.head()
```

```python
Mortality_Rate = new_cdc.loc[new_cdc["Question"] == "Premature mortality among adults aged 45-64 years", :]
Mortality_Rate.head()
```

```python
Overall_Mortality = Mortality_Rate.loc[Mortality_Rate["StratificationCategory1"] == "Overall",:]
Overall_Mortality.head()
```

```python
Mortality_Drop = Overall_Mortality.dropna()
Mortality_Drop.head()
```

```python
Number_Mortality_Drop = Mortality_Drop.loc[Mortality_Drop["DataValueType"] == "Number",:]
Number_Mortality_Drop.head()
```

```python
eleven_insurance = Crude_Insurance.loc[Crude_Insurance["YearEnd"] == 2011,:]
eleven_insurance_data = eleven_insurance["DataValue"].astype(float)
avg_eleven_insurance = eleven_insurance_data.mean()
avg_eleven_insurance
```

```python
twelve_insurance = Crude_Insurance.loc[Crude_Insurance["YearEnd"] == 2012,:]
twelve_insurance_data = twelve_insurance["DataValue"].astype(float)
avg_twelve_insurance = twelve_insurance_data.mean()
avg_twelve_insurance
```

```python
thirteen_insurance = Crude_Insurance.loc[Crude_Insurance["YearEnd"] == 2013,:]
thirteen_insurance_data = thirteen_insurance["DataValue"].astype(float)
avg_thirteen_insurance = thirteen_insurance_data.mean()
avg_thirteen_insurance
```

```python
fourteen_insurance = Crude_Insurance.loc[Crude_Insurance["YearEnd"] == 2014,:]
fourteen_insurance_data = fourteen_insurance["DataValue"].astype(float)
avg_fourteen_insurance = fourteen_insurance_data.mean()
avg_fourteen_insurance
```

```python
fifteen_insurance = Crude_Insurance.loc[Crude_Insurance["YearEnd"] == 2015,:]
fifteen_insurance_data = fifteen_insurance["DataValue"].astype(float)
avg_fifteen_insurance = fifteen_insurance_data.mean()
avg_fifteen_insurance
```

```python
eleven_mortality = Number_Mortality_Drop.loc[Number_Mortality_Drop["YearEnd"] == 2011,:]
eleven_mortality_data = eleven_mortality["DataValue"].astype(int)
avg_eleven_mortality = eleven_mortality_data.mean()
avg_eleven_mortality
```

```python
twelve_mortality = Number_Mortality_Drop.loc[Number_Mortality_Drop["YearEnd"] == 2012,:]
twelve_mortality_data = twelve_mortality["DataValue"].astype(int)
avg_twelve_mortality = twelve_mortality_data.mean()
avg_twelve_mortality
```

```python
thirteen_mortality = Number_Mortality_Drop.loc[Number_Mortality_Drop["YearEnd"] == 2013,:]
thirteen_mortality_data = thirteen_mortality["DataValue"].astype(int)
avg_thirteen_mortality = thirteen_mortality_data.mean()
avg_thirteen_mortality
```

```python
fourteen_mortality = Number_Mortality_Drop.loc[Number_Mortality_Drop["YearEnd"] == 2014,:]
fourteen_mortality_data = fourteen_mortality["DataValue"].astype(int)
avg_fourteen_mortality = fourteen_mortality_data.mean()
avg_fourteen_mortality
```

```python
Years = ["2011", "2012", "2013", "2014"]

Insurance_Averages = [avg_eleven_insurance, avg_twelve_insurance, avg_thirteen_insurance, avg_fourteen_insurance]

Mortality_Averages = [avg_eleven_mortality, avg_twelve_mortality, avg_thirteen_mortality, avg_fourteen_mortality]
```

```python
x = Years
y = Insurance_Averages 
plt.scatter(x, y)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
```

```python



plt.scatter(Years, Insurance_Averages, marker="o", facecolors="red", edgecolors="black", linewidth=1)
plt.title("Current lack of health insurance among adults aged 18-64 years")
plt.xlabel("Year")
plt.ylabel("Percentage of Uninsured")
plt.savefig("Resources/Lack_of_Insurance.png")
plt.show()
```
